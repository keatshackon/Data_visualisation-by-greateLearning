```python
import numpy as np

from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
y = x**4
```

```python
x
y
```

```python
x
```

```python
plt.plot(x,y,color='r',linestyle=":",linewidth = 2)
plt.title("Mr.Keats   VS   Time")
plt.ylabel("Mr.Keats")
plt.xlabel("Time")
```

## Plot Multiple Line In Same Graph

```python


y1 = 2 * x
y2 = 3 * x

plt.plot(x,y1,color='r',linestyle=':')
plt.plot(x,y2,color='g')
plt.title("Mr.Keats   VS   Time")
plt.ylabel("Mr.Keats")
plt.xlabel("Time")
plt.grid(True)
plt.show()
```

#  Bar Chat Plotting

```python
students = {"Mr.keats":92,"Salazar":90,"Sineboi":80,"Scientist":99,"Damayan":85}
```

```python
students
```

```python
names = list(students.keys())
marks = list(students.values())


plt.bar(names,marks,color='r')
plt.title("Students   VS   Marks")
plt.ylabel("Marks")
plt.xlabel("Students")
plt.show()
# to make horzontal bar chart use plt.barh(names,marks)
plt.barh(names,marks)
```

# Scatter Plotting

```python
import numpy as np

x = [3,23,34,45,45,95,15,32,44,62,67]
y = [85,4,5,56,7,61,87,67,17,29,23]
z = [33,45,51,49,79,69,37,67,37,39,11]

plt.scatter(x,y,color='r',marker="*",s=100)
plt.scatter(x,z,color='g',s=100)
plt.show()
```

# Sub-Plots In Scatter Plotting

```python
plt.subplot(1,2,1)
plt.scatter(x,y,color='r',marker="*",s=100)

plt.subplot(1,2,2)
plt.scatter(x,z,color='g',s=100)
```

# Pie-Chart Plotting

```python
fruits = ['Apple',"Orages","Grapes","Papaya","Pine Apple","Chiku","Bananas","Water Melon","Musk Melon"]
quantity  = [20,30,40,50,60,70,80,90,100]
```

```python
plt.pie(quantity,labels = fruits,autopct="%0.1f%%")
plt.show()
```

# DoughNut-Chart Plotting

```python
plt.pie(quantity,labels = fruits,radius=1)
plt.pie([1],colors=['w'],radius=0.5)
plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# fmri is in-built Data-Set in seabron

fmri = sns.load_dataset("fmri")
fmri.head()
fmri.tail()
```

```python
# To dimension of frmi data-set use shape() attribute
print(fmri.shape)

sns.lineplot(x = "timepoint",y="signal",data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",hue = "event",data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",hue = "event",data=fmri,style="event",markers=True)
plt.show()
```

```python
diamonds = pd.read_csv("diamonds.csv")
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
# to remove histograme from background
sns.distplot(diamonds['price'],hist = False)
plt.show()
```

```python
# to use different color use color attributes
sns.distplot(diamonds['price'],color = "purple")
plt.show()
```

```python
# to remove the frequency Curve use kde
sns.distplot(diamonds['price'],color = "purple",kde=False)
plt.show()
```

```python
sns.distplot(diamonds['price'],color = "purple",kde=False,bins=5)
plt.show()
```

```python
# vertical = true  to  rotate the curve into 90 degree
sns.distplot(diamonds['price'],color = "orange",bins=5,vertical = True)
plt.show()
```

# jointplot in seaborn


```python
iris = pd.read_csv("iris.csv")
sns.jointplot(x="SepalLengthCm",y="PetalLengthCm",data=iris)
plt.show()
```

```python
# to change color use color attribute
sns.jointplot(x="SepalLengthCm",y="PetalLengthCm",data=iris,color = "olive")
plt.show()
```

```python
# to draw thw regression line kind atributes
sns.jointplot(x="SepalLengthCm",y="PetalLengthCm",data=iris,kind = "reg")
plt.show()
```
